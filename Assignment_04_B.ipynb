```python
# ==========================================
# 1. INSTALL SYSTEM DEPENDENCIES
# ==========================================
# Install Tesseract OCR and silence the output
!sudo apt-get update -qq
!sudo apt-get install tesseract-ocr -qq
!pip install pytesseract Pillow -qq

# ==========================================
# 2. IMPORT LIBRARIES
# ==========================================
import pytesseract
from PIL import Image
from google.colab import files
import io

# ==========================================
# 3. UPLOAD AND PROCESS IMAGE
# ==========================================
print("👇 CLICK THE BUTTON BELOW TO UPLOAD YOUR IMAGE 👇")
print("-" * 50)

# This opens the browser upload dialog box
uploaded = files.upload()

print("-" * 50)

# Check if a file was actually uploaded
if not uploaded:
    print("❌ No file was uploaded. Please re-run the cell to try again.")
else:
    # CHANGED: .items() correctly unpacks both the filename and the content
    for filename, file_content in uploaded.items():
        print(f"📄 File successfully uploaded: {filename}")
        print("Running OCR extraction...\n")
        print("--- EXTRACTED TEXT START ---")

        try:
            # Load the image data directly from memory
            img = Image.open(io.BytesIO(file_content))

            # Extract text
            text = pytesseract.image_to_string(img)

            # Display results
            if text.strip():
                print(text)
            else:
                print("⚠️ OCR complete, but no text was detected in this image.")

        except Exception as e:
            print(f"❌ An error occurred while reading the image: {e}")

        print("--- EXTRACTED TEXT END ---")
```
